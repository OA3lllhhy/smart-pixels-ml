```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plot training and validation loss against epochs
csv_logger_path = "/home/submit/haoyun22/smart-pixels-ml/csvlogs/training_log_2ts5000.csv"
data = pd.read_csv(csv_logger_path)

# Print the latest epoch number
latest_epoch = data['epoch'].iloc[-1]
print(f"Latest epoch number: {latest_epoch}")

plt.figure(figsize=(10, 6))
plt.plot(data['epoch'], data['loss'], label='Training Loss')
plt.plot(data['epoch'], data['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss over Epochs 2 Time Slices 5000 Batch Size')
plt.legend()
plt.grid()
plt.show()
```
